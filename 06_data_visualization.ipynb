```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
# bar plot using seaborn
plt.figure(figsize=(8,4))
palette = {"First": "blue", "Second": "pink","Third": "green"}
sns.barplot(data=df, x="sex", y="fare",hue="class",palette=palette,capsize=.1)
ylabel="fare"
xlabel="sex"
plt.title("fare vs sex")
plt.savefig("fare vs sex.png")
plt.legend(loc="upper right", title="survived")
plt.show()
```

```python
# sactter plot
sns.scatterplot(data=df, x="age", y="fare",hue="class")
plt.figure(figsize=(10,5))
plt.title("fare of the passenger according to age")
plt.legend(loc="upper left", title="Class")
plt.xlabel="Age"
plt.ylabel="Fare"
plt.savefig("./output/fare according to age.png")
```

```python
df['survived']=df['survived'].map({0:'No',1:'Yes'})
```

```python
# scatter plot of figure size
plt.figure(figsize=(10,5))
palette={'No':'red','Yes':'green'}
sns.scatterplot(data=df, x="age", y="fare",hue="survived",palette=palette)
```

Interpertation
1.We have age at x-axis and fare at y-axis
2.The scatterplot shows the relationship between age and fare of titanic passengers
3.We further grouped the passengers into survived(green) and non-survived (red)
4.Two passengers have given highest fare 500 pounds and are in the survived category
5.The oldest passenger 80 years at titanic is also a survived one

```python
# Box plot
df['class'].unique()
```

```python
# style of a plot
sns.set_theme(style="darkgrid",font='calibri',font_scale=1)
palette={'No':'red','Yes':'green'}
sns.boxplot(data=df, x="class", y="fare",hue="survived",palette=palette)
plt.xlabel="class"
plt.ylabel="Fare"
plt.title("Fare according to class")
plt.savefig("./output/fare according to class.png")
plt.show()
```

```python
# Boxenplot
sns.boxenplot(data=df, x="class", y="fare",hue="survived",palette=palette)
plt.xlabel="class"
plt.ylabel="Fare"
plt.title("Fare according to class")
```

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(10,5)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='calibri')
# scatter plot size of figure size
plt.figure(figsize=(8,4))
palette={'No':'red','Yes':'green'}
sns.barplot(data=df, x='sex', y='age', hue='survived', palette=palette)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Barplot of Age, based on Gender and Age of Passengers")
# change legend position and title
plt.legend(loc='upper left', title='Survived')
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=45)
```
